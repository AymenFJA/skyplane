```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn-bright")


labels = ['aws:ap-northeast-2\n -> gcp:us-central1-a', 'gcp:europe-north1-a\n -> gcp:us-west4-a', 'aws:us-east-1\n -> gcp:us-west4-a']
gcp = [308, 105, 284]
skylark = [61.4, 57.4, 55.2]
skylark_instance = [16.3, 12.4, 21.1]

x = np.arange(len(labels))  # the label locations
width = 0.22  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, gcp, width, label='GCP Data Transfer')
rects2 = ax.bar(x, skylark, width, label='Skylark')
rects3 = ax.bar(x + width, skylark_instance, width, label='Skylark (overlay-only)')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Transfer Time (s)')
ax.set_title('Data Transfer Time', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_xlabel("Transfer Regions", fontsize=12, fontweight='bold')
ax.set_ylabel('Transfer Time (s)', fontsize=12, fontweight='bold')
ax.legend(loc='upper center')


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 1),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
